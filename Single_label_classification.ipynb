```python
from fastai.vision.all import *
```

```python
path = untar_data(URLs.PETS)
path.ls()
```

```python
files = get_image_files(path/'images')
len(files)
```

```python
def is_cat(x): return x[0].isupper()
```

```python
dls = ImageDataLoaders.from_name_func(path, files, is_cat, item_tfms=Resize(224))
```

```python
dls.show_batch()
```

```python
model = cnn_learner(dls, resnet34,metrics=accuracy)
model.fine_tune(10)
```

```python
model1 = cnn_learner(dls, resnet34,metrics=accuracy)
model1.fine_tune(1)
```

```python
model.predict(files[0])
```

```python
model.show_results()
```
